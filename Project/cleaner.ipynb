```python
import re
```

```python
pattern = r'[^a-zA-Z0-9ऀ-ॏ०-९\s*#]'
```

```python
def cleaner(textlist): # for removing punctuantions and other special stuff for proper evaluation
    output = []
    o2 = []
    textlist = textlist.split()
    for items in textlist:
      if '@' in items or 'http' in items:
        continue
      else:
        o2.append(items)
    o2 = ' '.join(o2)
    # textlist = ' '.join(sentence.split()[1:])
    for i in range(len(o2)):
        # temp = ' '.join(sentence.split()[1:])
        temp = re.sub(pattern, ' ', o2[i])
        # output.append(temp.split())
        output.append(temp)
    output = ''.join(output)
    output = re.sub(r'\s+', ' ', output)
    return output.strip()
```

```python
'@' in sentence.split()[0]
```

```python
cleaner(sentence)
```

```python
for i in range(len(train)):
  train['text'][i] = cleaner(train['text'][i])
for i in range(len(test)):
  test['text'][i] = cleaner(test['text'][i])
for i in range(len(val)):
  val['text'][i] = cleaner(val['text'][i])
```

```python
train.to_excel('/content/gdrive/MyDrive/cs689/project/data/clean2_train.xlsx')

test.to_excel('/content/gdrive/MyDrive/cs689/project/data/clean2_test.xlsx')

val.to_excel('/content/gdrive/MyDrive/cs689/project/data/clean2_valid.xlsx')
```
